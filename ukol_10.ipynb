```python
import pandas
```

```python
liberec = pandas.read_csv("zam_liberec.csv")
plzen = pandas.read_csv("zam_liberec.csv")
praha = pandas.read_csv("zam_praha.csv")
```

```python
liberec["mesto"] = "Liberec"
plzen["mesto"] = "Plzeň"
praha["mesto"] = "Praha"
liberec.head(5)
```

```python
zamestnanci = pandas.concat([liberec, plzen, praha], ignore_index=True)
zamestnanci
```

```python
platy = pandas.read_csv("platy_2021_02.csv", sep=",")
platy
```

```python
propojeny_df = pandas.merge(zamestnanci, platy, on=["cislo_zamestnance"])
propojeny_df
```

```python
propojeny_df.groupby("mesto")["plat"].mean()
```
